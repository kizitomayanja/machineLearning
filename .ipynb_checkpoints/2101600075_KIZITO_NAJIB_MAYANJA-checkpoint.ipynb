```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('top2018.csv')
```

```python
cat_cols = data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical variables")
print(cat_cols)
print("Numerical variables")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(),2))
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1,2,2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
fig, axes = plt.subplots(2,1, figsize = (24,24))
fig.suptitle('Bar plot for song names and artists')
sns.countplot(ax=axes[0], x = 'name', data = data, color = 'red', order = data['name'].value_counts().index)
sns.countplot(ax=axes[1], x = 'artists',data = data, color = 'red', order = data['artists'].value_counts().index)
axes[0].tick_params(labelrotation=90)
axes[1].tick_params(labelrotation=90)
```

```python
def log_trans(data,col):
    for colname in col:
        if(data[colname] == 1.0).all():
            data[colname+'_log'] = np.log(data[colname]+1e-8)
        else:
            data[colname+'_log']=np.log(data[colname])
    data.info()
```

```python
log_trans(data,['time_signature','instrumentalness','duration_ms'])
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(['instrumentalness','duration_ms','time_signature'],axis=1))
plt.show()
```

```python
fig, axarr = plt.subplots(10, 1, figsize=(24, 40))
data.groupby('artists')['danceability'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0], fontsize=12)
axarr[0].set_title("Artists Vs Danceability", fontsize=18)
data.groupby('artists')['energy'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1], fontsize=12)
axarr[1].set_title("Artists Vs Energy", fontsize=18)
data.groupby('artists')['loudness'].mean().sort_values(ascending=False).plot.bar(ax=axarr[2], fontsize=12)
axarr[2].set_title("Artists Vs Loudness", fontsize=18)
data.groupby('artists')['acousticness'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3], fontsize=12)
axarr[3].set_title("Artists Vs Acousticness", fontsize=18)
data.groupby('artists')['key'].mean().sort_values(ascending=False).plot.bar(ax=axarr[4], fontsize=12)
axarr[4].set_title("Artists Vs Key", fontsize=18)
data.groupby('artists')['duration_ms'].mean().sort_values(ascending=False).plot.bar(ax=axarr[5], fontsize=12)
axarr[5].set_title("Artists Vs Duration", fontsize=18)
data.groupby('artists')['instrumentalness_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[6], fontsize=12)
axarr[6].set_title("Artists Vs Instrumentalness", fontsize=18)
data.groupby('artists')['liveness'].mean().sort_values(ascending=False).plot.bar(ax=axarr[7], fontsize=12)
axarr[7].set_title("Artists Vs Liveness", fontsize=18)
data.groupby('artists')['tempo'].mean().sort_values(ascending=False).plot.bar(ax=axarr[8], fontsize=12)
axarr[8].set_title("Artists Vs Tempo", fontsize=18)
data.groupby('artists')['speechiness'].mean().sort_values(ascending=False).plot.bar(ax=axarr[9], fontsize=12)
axarr[9].set_title("Artists Vs Speechiness", fontsize=18)
plt.subplots_adjust(hspace=3.0)
plt.subplots_adjust(wspace=2.5)
sns.despine()
```
